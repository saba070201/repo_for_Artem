## Рекурсия


### Реализация

```python
def factorial_recursive(n):
    k=n
    if n == 1 or n==0:
        return n
    else:
        
        return n*factorial_recursive(k-1) # 3 * (2 * 1 )
res=factorial_recursive(3)
print(res)
```

## Бинарный поиск

### Реализация

```python
def simple_search(arr,value):
    j=1
    for i in range(len(arr)):
        print(j)
        j+=1
        if arr[i]==value:
            return i
def binary_search(arr,value):
    low_index=0
    high_index=(len(arr))-1
    i=1
    while low_index<=high_index:
        print(i)
        i+=1
        mid_index=(low_index+high_index)//2
        if arr[mid_index]==value:
            return mid_index
        if arr[mid_index]>value:
            high_index=mid_index-1
        else:
            low_index=mid_index+1
    return None
arr=list(range(100))

print(binary_search(arr,99))
```

## Сортировки

### Пузурьком 

#### Реализация

```python
def bubble(array):
    for i in range(len(array)-1):
        for j in range(len(array)-i-1):
            if array[j] > array[j+1]:
                buff = array[j]
                array[j] = array[j+1]
                array[j+1] = buff
     
arr=[random.randint(1,1000) for i in range(100000) ]

bubble(arr)
```

### Расческой

#### Реализация 

```python
def comb(array):
    gap = len(array)
    swaps = True
    while gap > 1 or swaps:
        gap = int(gap // 1.257)  
        swaps = False
        for i in range(len(array) - gap):
            j = i + gap
            if array[i] > array[j]:
                array[i], array[j] = array[j], array[i]
                swaps = True
            
arr=[random.randint(1,1000) for i in range(100000)]
print(arr)
comb(arr)
```

### Быстрая

```python
def quicksort(array):
    if len(array) <= 1:
        return array
    else:
        q = random.choice(array)
        print(q)
        left = []
        midle = []
        right = []
        for elem in array:
            if elem < q:
                left.append(elem) 
            elif elem > q: 
                right.append(elem) 
            else: 
                midle.append(elem)
        return quicksort(left) + midle + quicksort(right)
# quicksort([random.randint(1,10) for i in range(10)])
print([random.randint(1,10) for i in range(100)])
```

### Поиск в глубину

```python
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)

    print(start)

    for next in graph[start] - visited:
        dfs(graph, next, visited)
    return visited



```

```python
def bfs(graph, start):
    visited, queue = set(), [start]
    while queue:
        vertex = queue.pop(0)
        if vertex not in visited:
            visited.add(vertex)
            queue.extend(graph[vertex] - visited)
    return visited
```

### Алгоритм флойда

```python

import math


def get_path(P, u, v):
    path = [u]
    while u != v:
        u = P[u][v]
        path.append(u)

    return path


V = [[0, 2, math.inf, 3, 1, math.inf, math.inf, 10],
     [2, 0, 4, math.inf, math.inf, math.inf, math.inf, math.inf],
     [math.inf, 4, 0, math.inf, math.inf, math.inf, math.inf, 3],
     [3, math.inf, math.inf, 0, math.inf, math.inf, math.inf, 8],
     [1, math.inf, math.inf, math.inf, 0, 2, math.inf, math.inf],
     [math.inf, math.inf, math.inf, math.inf, 2, 0, 3, math.inf],
     [math.inf, math.inf, math.inf, math.inf, math.inf, 3, 0, 1],
     [10, math.inf, 3, 8, math.inf, math.inf, 1, 0],
]

N = len(V)                       # число вершин в графе
P = [[v for v in range(N)] for u in range(N)]       # начальный список предыдущих вершин для поиска кратчайших маршрутов

for k in range(N):
    for i in range(N):
        for j in range(N):
            d = V[i][k] + V[k][j]
            if V[i][j] > d:
                V[i][j] = d
                P[i][j] = k     # номер промежуточной вершины при движении от i к j

# нумерацця вершин начинается с нуля
start = 1
end = 4
print(get_path(P, start, end))
```
